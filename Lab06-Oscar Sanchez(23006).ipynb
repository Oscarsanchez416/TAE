```python
# analizar el crecimiento de la poblacion de Afghanistan y honduras
```

```python
data = pd.read_csv("countries.csv")
```

```python
data
```

```python
Afghanistan = data[data.country == "Afghanistan"]
```

```python
Afghanistan
```

```python
honduras = data[data.country == "Honduras"]
```

```python
honduras
```

```python
plt.plot(Afghanistan.year, Afghanistan.population/10**6)
plt.plot(honduras.year, honduras.population/10**6)
plt.legend(["Afghanistan", "Honduras"])
plt.xlabel("Año")
plt.ylabel("Poblacion (millones)")
plt.show()
```

```python
Afghanistan.population
```

```python
Afghanistan.population / Afghanistan.population.iloc[0]
```

```python
Afghanistan.population / Afghanistan.population.iloc[0]*100
```

```python
plt.plot(Afghanistan.year, Afghanistan.population / Afghanistan.population.iloc[0]*100)
plt.plot(honduras.year, honduras.population / honduras.population.iloc[0]*100)
plt.legend(["Afghanistan", "Honduras"])
plt.xlabel("Año")
plt.ylabel("Crecimiento de la poblacion(1er año 100)")
plt.show()
```
