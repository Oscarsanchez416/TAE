```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data = pd.read_csv("TEA-Data.csv")
data
```

```python
fig, ax = plt.subplots(figsize=(10,5)) 
sns.heatmap(data.groupby(['Day','Time_Hours'])['Cost_of_Goods_Sold'].sum()
              .unstack('Time_Hours',fill_value=0), annot=True, fmt='g', cmap='Greens')
```
